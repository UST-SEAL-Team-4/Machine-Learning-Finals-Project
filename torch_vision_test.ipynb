```python
import numpy as np
import nibabel as nib
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def plot_3d_CMB(image_data, threshold=None):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    
    x, y, z = np.where(image_data > threshold) if threshold is not None else np.where(image_data > 0)
    ax.scatter(x, y, z, zdir='z', c='red', marker='.')
    
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')

    plt.show()


def plot_3d_MRI(image_data, threshold=None):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    
    x, y, z = np.where(image_data > threshold) if threshold is not None else np.where(image_data > 0)
    ax.scatter(x, y, z, zdir='z', c='blue', marker='.', alpha=0.010)
    
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')

    plt.show()

# Load MRI scans
mri_img = nib.load('Task2/sub-101/sub-101_space-T2S_desc-masked_T2S.nii.gz')
mri_data = mri_img.get_fdata()

# Load segmented CMBs
cmb_img = nib.load('Task2/sub-101/sub-101_space-T2S_CMB.nii.gz')
cmb_data = cmb_img.get_fdata()

# Plot 3D MRI
plot_3d_MRI(mri_data)

# Plot 3D CMB
plot_3d_CMB(cmb_data)
```

```python
import numpy as np
import nibabel as nib
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
def plot_3d_CMB_MRI(CMB, MRI,threshold=None):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

    x, y, z = np.where(MRI > threshold) if threshold is not None else np.where(MRI > 0)
    ax.scatter(x, y, z, zdir='z', c='blue', marker='.', alpha=0.005)
    x, y, z = np.where(CMB > threshold) if threshold is not None else np.where(CMB > 0)
    ax.scatter(x, y, z, zdir='z', c='red', marker='.')


# Load MRI scans
mri_img = nib.load('Task2/sub-101/sub-101_space-T2S_desc-masked_T2S.nii.gz')
mri_data = mri_img.get_fdata()

# Load segmented CMBs
cmb_img = nib.load('Task2/sub-101/sub-101_space-T2S_CMB.nii.gz')
cmb_data = cmb_img.get_fdata()

# Plot 3D of MRI and CMB combined
plot_3d_CMB_MRI(cmb_data, mri_data)
```
