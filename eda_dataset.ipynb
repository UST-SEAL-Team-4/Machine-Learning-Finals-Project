### Create DataFrame and get the number of slices per case


```python
import pandas as pd


def get_slice_count(image_paths, cohort_name):
    data = []
    for index, img_path in enumerate(image_paths):
        img = nib.load(img_path)
        num_slices = img.shape[2]
        case_id = index + 1
        data.append({
            'Cohort Number': cohort_name,
            'Case ID': case_id,
            'Number of Slices': num_slices
        })
    return data


data_cohort_one = get_slice_count(cohort_one_imgs, 1)
data_cohort_two = get_slice_count(cohort_two_imgs, 2)
data_cohort_three = get_slice_count(cohort_three_imgs, 3)

df = pd.DataFrame(data_cohort_one + data_cohort_two + data_cohort_three)
df
```

### Check if all the slices per case have consistent slice thickness


```python
def get_image_spacing(images, cohort_number):
    data = []
    for index, img_path in enumerate(images):
        img = nib.load(img_path)
        spacings = img.header.get_zooms()

        num_slices = img.shape[2]
        case_id = index + 1

        for slice_index in range(num_slices):
            data.append({
                "Cohort Number": cohort_number,
                "Case ID": case_id,
                "Slice Index": slice_index + 1,
                "Slice Thickness (mm)": spacings[2]
            })

    return data


cohort_one_spacing = get_image_spacing(cohort_one_imgs, 1)
cohort_two_spacing = get_image_spacing(cohort_two_imgs, 2)
cohort_three_spacing = get_image_spacing(cohort_three_imgs, 3)

df_slice_thickness = pd.DataFrame(
    cohort_one_spacing + cohort_two_spacing + cohort_three_spacing)
df_slice_thickness
```

### Check the number of unique slice thickness per case


```python
unique_thickness_per_case = df_slice_thickness.groupby(
    ['Cohort Number', 'Case ID'])['Slice Thickness (mm)'].nunique()

unique_thickness_per_case = unique_thickness_per_case.reset_index()
unique_thickness_per_case.rename(
    columns={'Slice Thickness (mm)': 'Unique Slice Thickness Count'}, inplace=True)

unique_thickness_per_case
```

### Add Slice Thickness in the DataFrame


```python
# Since all cases have consistent slice thickness, we will add this information in the dataframe
def get_slice_thickness(images, cohort_number, df):
    for index, img_path in enumerate(images):
        img = nib.load(img_path)
        spacings = img.header.get_zooms()
        case_id = index + 1

        df.loc[(df['Cohort Number'] == cohort_number) & (
            df['Case ID'] == case_id), 'Slice Thickness (mm)'] = spacings[2]


get_slice_thickness(cohort_one_imgs, 1, df)
get_slice_thickness(cohort_two_imgs, 2, df)
get_slice_thickness(cohort_three_imgs, 3, df)

df
```

### Check number of CMBs per case


```python
import numpy as np
import cv2


def extract_bounding_boxes(image_data):
    contours, _ = cv2.findContours(image_data.astype(
        np.uint8), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    return len(contours)


def count_cmb_per_image(segmented_images):
    cmb_counts = []
    for img_path in segmented_images:
        img = nib.load(img_path)
        data = img.get_fdata()
        slice_cmb_counts = [extract_bounding_boxes(
            (data[:, :, i] > 0).astype(np.uint8)) for i in range(data.shape[2])]
        total_cmb_count = sum(slice_cmb_counts)
        cmb_counts.append(total_cmb_count)
    return cmb_counts


cohort_one_cmb_counts = count_cmb_per_image(cohort_one_segmented_imgs)
cohort_two_cmb_counts = count_cmb_per_image(cohort_two_segmented_imgs)
cohort_three_cmb_counts = count_cmb_per_image(cohort_three_segmented_imgs)

print("CMB Counts in Cohort 1:", cohort_one_cmb_counts)
print("CMB Counts in Cohort 2:", cohort_two_cmb_counts)
print("CMB Counts in Cohort 3:", cohort_three_cmb_counts)
```

### Add the Slice Shape in the DataFrame


```python
df
```

```python
def add_slice_shape(df, image_paths, cohort_number):
    if 'Slice Height' not in df.columns:
        df['Slice Height'] = None
    if 'Slice Width' not in df.columns:
        df['Slice Width'] = None

    for index, img_path in enumerate(image_paths):
        img = nib.load(img_path)
        case_id = index + 1

        row_index = (df['Cohort Number'] == cohort_number) & (
            df['Case ID'] == case_id)
        df.loc[row_index, 'Slice Height'] = img.shape[0]
        df.loc[row_index, 'Slice Width'] = img.shape[1]


add_slice_shape(df, cohort_one_imgs, 1)
add_slice_shape(df, cohort_two_imgs, 2)
add_slice_shape(df, cohort_three_imgs, 3)

df
```
